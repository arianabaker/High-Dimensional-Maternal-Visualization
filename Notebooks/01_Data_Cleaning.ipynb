## *Categorical Variables*

````python
## ----- Binary ----- ##
bincat_vars = ['rf_pdiab', 'rf_gdiab', 'rf_phype', 'rf_ghype', 'rf_ehype', 'rf_ppterm',
                       'mm_mtr', 'mm_plac', 'mm_rupt', 'mm_uhyst', 'mm_aicu','ip_gon', 'ip_syph',
                       'ip_chlam', 'ip_hepb', 'ip_hepc', 'ob_ecvs', 'ob_ecvf', 'ld_indl',
                       'ld_augm', 'ld_ster', 'ld_antb', 'ld_chor', 'ld_anes', 'me_trial',
                       'ab_aven1', 'ab_aven6', 'ab_nicu', 'ab_surf','ab_anti', 'ab_seiz','ca_anen',
                       'ca_mnsb', 'ca_cchd', 'ca_cdh','ca_omph','ca_gast', 'ca_limb', 'ca_cleft',
                       'ca_clpal', 'ca_hypo', 'wic']

## ----- Nominal ----- ##
nomcat_vars = ['mracehisp', 'me_pres', 'me_rout', 'rdmeth_rec']

## ----- Ordinal ----- ##'
ordcat_vars = ['meduc', 'lbo_rec', 'tbo_rec', 'precare']

print('Binary categorical variable count:', len(bincat_vars))
print('Nominal categorical variable count:', len(nomcat_vars))
````

### *Binary Encoding*

````python
for col in bincat_vars: # normalize text
  raw_data[col] = raw_data[col].astype(str).str.strip().str.upper()

for col in bincat_vars: # convert to 0/1
  raw_data[col] = (raw_data[col] == 'Y').astype(int)
````

# ***Clustering Data***

## *Create Cluster Dataset*

````python
# Set cluster data using cluster variables only
cluster_data = raw_data[cluster_vars].copy()

print(f'Shape of Cluster Dataset:', cluster_data.shape)
cluster_data.head()
````

````python
# Remove Inferitlity/Assist. Technology Variables
cluster_data = cluster_data.drop(columns=['rf_inftr', 'rf_artec', 'rf_fedrg'])
print(f'Shape of Cluster Dataset:', cluster_data.shape)
cluster_data.head()
````

The features pertaining to infertility treatements/drugs, and assisted reproductive technology will not be included as pre-delivery characteristics. They will be excluded because of a partial state-level reporting issues, where the information was inputted incorrectly-- hence vast missing data. However, there is no way to verify which rows came from the affected states. All three columns will be removed.

----

Note this public health data issue in report.

````python
# missingness
cluster_missing = cluster_data.isna().mean()*100
cluster_missing.sort_values(ascending = False)
````

## *Imputing NaN Data*

````python
# Median Impuation
med_impute_vars = disnum_vars + contnum_vars + ordcat_vars
med_impute_vars = [col for col in med_impute_vars if col in cluster_data.columns]

for col in med_impute_vars:
  cluster_data[col] = cluster_data[col].fillna(cluster_data[col].median())

# Verify
cluster_data[med_impute_vars].isna().sum()
````

## *One-hot Encoding*


````python
# Nominal Variable (Race) -- create 'Unknown' for one-hot encoding
cluster_data['mracehisp'] = cluster_data['mracehisp'].fillna('Unknown')

# One-hot encoding for Nominal variable
cluster_data = pd.get_dummies(cluster_data, columns = ['mracehisp'], drop_first = False)

# Convert true/false to 1/0
race_bool_cols = [col for col in cluster_data.columns if 'mracehisp_' in col]
cluster_data[race_bool_cols] = cluster_data[race_bool_cols].astype(int)

# Verify
cluster_data[race_bool_cols].head()
````
